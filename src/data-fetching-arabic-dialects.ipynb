```python
import pandas as pd

df = pd.read_csv('../input/arabic-dialects-ids-dataset/dialect_dataset.csv')
```

```python
df
```

```python
df['id'] = df['id'].apply(str)
ids = df['id'].tolist()
```

```python
import requests
from tqdm.auto import tqdm
step = 1000
responses = dict()
progress_bar = tqdm(range(0, len(df), step))
for i in range(0, len(df), step):
    response = requests.post('https://recruitment.aimtechnologies.co/ai-tasks', json = ids[i:i+step])
    responses.update(response.json())
    progress_bar.update()
```

```python
df_response = pd.DataFrame(responses.items(), columns=['id', 'text'])
df_response.tail()
```

```python
data = df.set_index('id').join(df_response.set_index('id'))
data.tail()
```

```python
data.to_csv('aradic_dialects.csv')
```
